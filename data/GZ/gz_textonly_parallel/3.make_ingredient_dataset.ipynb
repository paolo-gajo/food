```python
# build ingredient dataset

import os
import re
import json

def convert_fractions_to_plain(text):
    fraction_mapping = {
        '½': '1/2', '¼': '1/4', '¾': '3/4',
        '⅓': '1/3', '⅔': '2/3', '⅕': '1/5',
        '⅖': '2/5', '⅗': '3/5', '⅘': '4/5',
        '⅙': '1/6', '⅚': '5/6', '⅛': '1/8',
        '⅜': '3/8', '⅝': '5/8', '⅞': '7/8',
    }

    # Use a regular expression to find all fraction characters
    fraction_pattern = re.compile('|'.join(re.escape(key) for key in fraction_mapping.keys()))

    # Replace each fraction character with its corresponding string
    for fraction_char, replacement in fraction_mapping.items():
        text = text.replace(fraction_char, replacement)
    
    # Replace the '⁄' character with '/'
    text = text.replace("⁄", "/")

    return text


root = '/home/pgajo/working/food/data/gz_textonly_parallel/gf_parallel'
path_en = os.path.join(root, 'gf_en')
path_it = os.path.join(root, 'gf_it')

recipe_ingredient_list = []
new_dir_order = []

for dir in os.listdir(path_en):
    new_dir_order.append(int(dir.split("_")[-1]))

new_dir_order.sort()
# print(new_dir_order[:3])
dataset = []

for dir in new_dir_order:
    # print('gf_it_'+str(dir))
    
    recipe_path_en = os.path.join(path_en, 'gf_en_'+str(dir))
    recipe_path_it = os.path.join(path_it, 'gf_it_'+str(dir))

    # print(recipe_path_en)
    # print(os.listdir(recipe_path_en))
    # print(recipe_path_it)
    # print(os.listdir(recipe_path_it))

    with open(os.path.join(recipe_path_en, 'title.txt'), 'r', encoding='utf8') as f: 
        content_title_en = f.read()
        title_en = content_title_en.split('\n')[0]
        url_en = content_title_en.split('\n')[1]

    with open(os.path.join(recipe_path_it, 'title.txt'), 'r', encoding='utf8') as f:
        content_title_it = f.read()
        title_it = content_title_it.split('\n')[0]
        url_it = content_title_it.split('\n')[1]

    with open(os.path.join(recipe_path_en, 'presentation.txt'), 'r', encoding='utf8') as f:
        presentation_en = f.read()

    with open(os.path.join(recipe_path_it, 'presentation.txt'), 'r', encoding='utf8') as f:
        presentation_it = f.read()

    with open(os.path.join(recipe_path_en, 'ingredients.txt'), 'r', encoding='utf8') as f:
        lines = f.readlines()
        ingredients_en = ';'.join([line.strip() for line in lines])
        ingredients_en = convert_fractions_to_plain(re.sub(r'\s+', ' ', ingredients_en))

    with open(os.path.join(recipe_path_it, 'ingredients.txt'), 'r', encoding='utf8') as f:
        lines = f.readlines()
        ingredients_it = ';'.join([line.strip() for line in lines])
        ingredients_it = convert_fractions_to_plain(re.sub(r'\s+', ' ', ingredients_it))
    
    with open(os.path.join(recipe_path_en, 'preparation.txt'), 'r', encoding='utf8') as f:
        preparation_en = f.read()

    with open(os.path.join(recipe_path_it, 'preparation.txt'), 'r', encoding='utf8') as f:
        preparation_it = f.read()

    entry = {
    'id': dir,
    'url_en': url_en,
    'title_en': title_en,
    'presentation_en': presentation_en,
    'ingredients_en': ingredients_en,
    'preparation_en': preparation_en,
    'url_it': url_it,
    'title_it': title_it,
    'presentation_it': presentation_it,
    'ingredients_it': ingredients_it,
    'preparation_it': preparation_it,
    }

    for key in entry.keys():
        if isinstance(entry[key], str):
            entry[key] = entry[key].replace(u'\xa0', u' ')

    # print(entry)

    dataset.append(entry)

with open('gz_recipes.json', 'w', encoding='utf8') as f:
    json.dump(dataset, f, ensure_ascii=False)
```

```python
def convert_fractions_to_plain(text):
    fraction_mapping = {
        '½': '1/2', '¼': '1/4', '¾': '3/4',
        '⅓': '1/3', '⅔': '2/3', '⅕': '1/5',
        '⅖': '2/5', '⅗': '3/5', '⅘': '4/5',
        '⅙': '1/6', '⅚': '5/6', '⅛': '1/8',
        '⅜': '3/8', '⅝': '5/8', '⅞': '7/8',
    }
    fraction_pattern = re.compile('|'.join(re.escape(key) for key in fraction_mapping.keys()))

    for match in re.finditer(fraction_pattern, text):
        fraction_index = match.start()
        fraction = match.group()
        # print(fraction)
        three_char_fraction = fraction_mapping[fraction]
        # print(three_char_fraction)
        text = text[:fraction_index] + three_char_fraction + text[fraction_index + 1:]
    
    text = text.replace("⁄", "/")

    return text

```
